```python
# load saved friedman docs data pulled from nyt api
all_docs = cPickle.load(open('data/raw/all_friedman_docs.pkl', 'r'))
print 'len(all_docs) =', len(all_docs)
all_docs[0]
```

```python
# shuffle the docs for maximum obfuscation
random.shuffle(all_docs)
```

```python
# open full text output file
#out_file = open('/Users/bjdewilde/Desktop/data_science/projects/thomas_friedman_nlp/data/raw/corpus.txt', 'w')
out_file = open('/Users/bjdewilde/Desktop/data_science/projects/thomas_friedman_nlp/data/raw/corpus.txt', 'a')

# set requests session
session = requests.Session()

# iterate!
one_percent = max(int(0.01*len(all_docs)), 1)
for i,doc in enumerate(all_docs):
    
    if i < 6350: continue
    
    # mind the server
    time.sleep(random.random())
    
    # get a new IP address
    get_new_id()
    # check it?
#     ip_test = session.get('http://ifconfig.me/ip')
#     if ip_test.status_code == 200:
#         print 'IP =', ip_test.content
    
    # handy counter
    if i % 10 == 0:
        print '\n*i* =', i
    
    # scrape the article text
    url = doc.get('web_url')
    if not url:
        print '**WARNING: No url!'
        continue
    print 'url =', url
    
    text = hi_utilities.extract_article_text_from_url(url,
                                                      session=session,
                                                      host='www.nytimes.com',
                                                      referrer='http://topics.nytimes.com/top/opinion/editorialsandoped/oped/columnists/thomaslfriedman/index.html')
    if not text:
        print '**WARNING: No text!'
        continue
    elif text.startswith('Page Not Found'):
        print '**WARNING: Page Not Found!'
        continue
    if isinstance(text, unicode):
        text = unidecode(text)
    
    # add text to doc
    doc['full_text'] = text
    out_file.write(text+'\n')

out_file.close()
```

```python
friedman_docs = []
for doc in docs:
    if doc.get('full_text') and doc.get('byline') and doc['byline'].get('person'):
        for person in doc['byline']['person']:
            if person and person.get('lastname') and person['lastname'] == 'FRIEDMAN':
                friedman_docs.append(doc)
                break
print 'n_friedman_docs =', len(friedman_docs)
```

```python
df = pd.DataFrame(friedman_docs)
df
```

```python
df['type_of_material'].value_counts()
```

```python
list(df['keywords'])[:10]
```

```python
from collections import Counter
glocations = []; persons = []; subjects = []; organizations = []
for doc in friedman_docs:
    if not doc.get('keywords'):
        continue
    for keyword in doc['keywords']:
        if keyword['name'] == 'glocations':
            glocations.append(keyword.get('value'))
        elif keyword['name'] == 'persons':
            persons.append(keyword.get('value'))
        elif keyword['name'] == 'subject':
            subjects.append(keyword.get('value'))
        elif keyword['name'] == 'organizations':
            organizations.append(keyword.get('value'))
glocations_counter = Counter(glocations)
persons_counter = Counter(persons)
subjects_counter = Counter(subjects)
organizations_counter = Counter(organizations)
```

```python
glocations_counter.most_common(10)
```

```python
persons_counter.most_common(10)
```

```python
for s in subjects_counter.most_common(10):
    print s[0], '('+str(s[1])+')'
```

```python
organizations_counter.most_common(10)
```

```python
from dateutil.parser import parse as parse_date
import matplotlib.pyplot as plt
import pandas as pd
import prettyplotlib as ppl

oped_counts = pd.groupby(df[df['type_of_material'] == 'Op-Ed'], by='y')['full_text'].count()
news_counts = pd.groupby(df[df['type_of_material'] == 'News'], by='y')['full_text'].count()
other_counts = pd.groupby(df[~df['type_of_material'].isin(['News', 'Op-Ed'])], by='y')['full_text'].count()

fig = plt.figure(figsize=(12,6), dpi=100)
suptitle = fig.suptitle('NYT Articles by Thomas L. Friedman', fontsize=18)

ax = fig.add_subplot(1,1,1)
ppl.plot(ax, news_counts.index, news_counts.values,
         linewidth=3, marker='o', markersize=6, markeredgewidth=0.0,
         label='News')
ppl.plot(ax, oped_counts.index, oped_counts.values,
         linewidth=3, marker='o', markersize=6, markeredgewidth=0.0,
         label='Op-Eds')
ppl.plot(ax, other_counts.index, other_counts.values,
         linewidth=3, marker='o', markersize=6, markeredgewidth=0.0,
         label='Other')

xlab = ax.set_xlabel('Year', fontsize=14)
ylab = ax.set_ylabel('Number of Articles', fontsize=14)
ylim = ax.set_ylim((0,ax.get_ylim()[1]+20))

leg = plt.legend(loc='upper right', frameon=False, fontsize=14)

plt.tight_layout()

fig.savefig('/Users/bjdewilde/Desktop/data_science/projects/thomas_friedman_nlp/viz/article_counts_by_type_over_time.png', dpi=300)
```

### Sample Query for Blog Post

```python
import yaml
my_api_key = config = yaml.load(open('config.yaml').read())['nyt_api_key']
```

```python
from pprint import pprint
import requests

#my_api_key = ####

# API call parameters
fields = ['web_url', 'snippet', 'lead_paragraph', 'abstract',
          'print_page', 'blog', 'source', 'multimedia', 'headline',
          'keywords', 'pub_date', 'document_type', 'news_desk',
          'byline', 'type_of_material', '_id', 'word_count']
facet_fields = ['source', 'section_name', 'document_type', 'type_of_material', 'day_of_week']

resp = requests.get('http://api.nytimes.com/svc/search/v2/articlesearch.json',
                    params={'q': 'Thomas L. Friedman',
                            'page': 0,
                            'sort': 'newest',
                            'fl': ','.join(fields),
                            'facet_field': ','.join(facet_fields),
                            'facet_filter': 'true',
                            'api-key': my_api_key},
                    )
pprint(resp.json())
```
